## Tasks


###  Task 01 Load the data merge the data appropriately using pandas

```python
#Becaue product id is common in both dataframes and prodcut can be in different categories
merge_df = pd.merge(sales_df, product_df, how='inner', on='product_id')
```

```python
merge_df
```

### Task 02 Calculate the total revenue per category per month

```python
# Convert date column to datetime if not already
merge_df["date"] = pd.to_datetime(merge_df["date"])

# Extract month-year format
merge_df["month"] = merge_df["date"].dt.to_period("M")  # '2025-01', '2025-02', etc.

# Calculate revenue
merge_df["revenue"] = merge_df["quantity"] * merge_df["unit_price"]

# Group by month and category, then sum revenue
monthly_revenue = merge_df.groupby(["month", "category"])["revenue"].sum().reset_index()

```

```python

monthly_revenue
```

### Task 03 Find the top 3 products by revenue for each category

```python
# Calculate the revenue per product for each category
category_product_revenue = merge_df.groupby(['category', 'product_name'])['revenue'].sum().reset_index()

# Sort by category and revenue in descending order
category_product_revenue_sorted = category_product_revenue.sort_values(['category', 'revenue'], ascending=[True, False])

# Get the top 3 products by revenue for each category
top_3_products = category_product_revenue_sorted.groupby('category').head(3).reset_index(drop=True)

# Display the result with only top 3 products for each category
top_3_products
```

```python
import matplotlib.pyplot as plt

# Calculate the total revenue for the top 3 products in each category
category_revenue = top_3_products.groupby('category')['revenue'].sum().reset_index()

# Plot the total revenue for each category
plt.figure(figsize=(12, 6))
plt.bar(category_revenue['category'], category_revenue['revenue'], color='skyblue')

# Customize the plot
plt.title("Total Revenue from Top 3 Products by Category")
plt.xlabel("Category")
plt.ylabel("Total Revenue")
plt.xticks(rotation=45, ha='right')
plt.tight_layout()

# Show the plot
plt.show()
```

### 04 Find the product with the most profits (total revenues - total cost)


```python
# Calculate the cost (quantity * unit_cost)
merge_df['cost'] = merge_df['quantity'] * merge_df['unit_cost']

# Calculate profit (revenue - cost)
merge_df['profit'] = merge_df['revenue'] - merge_df['cost']

# Find the product with the most profit
product_most_profit = merge_df.groupby('product_name')['profit'].sum().reset_index()

# Sort by profit in descending order
product_most_profit_sorted = product_most_profit.sort_values('profit', ascending=False)

# Display the product with the most profit
product_most_profit_sorted.head(1)
```

### 05. Explain what strategy you would use to find products that have been performing particularly well or poorly in the last day

```python
# Calculate the daily revenue
merge_df['daily_revenue'] = merge_df['quantity'] * merge_df['unit_price']

# Set the date column to datetime if not already
merge_df['date'] = pd.to_datetime(merge_df['date'])

# Calculate the rolling 7-day average revenue for each product
merge_df['7_day_avg'] = merge_df.groupby('product_name')['daily_revenue'].rolling(window=7, min_periods=1).mean().reset_index(level=0, drop=True)

# Find the last day of data (today)
last_day_data = merge_df[merge_df['date'] == merge_df['date'].max()].copy()  

# Modify the copied DataFrame
last_day_data['revenue_diff'] = last_day_data['daily_revenue'] - last_day_data['7_day_avg']
last_day_data['performance'] = last_day_data['revenue_diff'].apply(lambda x: 'Spike' if x > 0 else 'Drop')

last_day_data[['product_name', 'daily_revenue', '7_day_avg', 'revenue_diff', 'performance']]
```
